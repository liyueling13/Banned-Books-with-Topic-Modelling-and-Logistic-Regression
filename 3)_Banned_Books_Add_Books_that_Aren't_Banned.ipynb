## Data cleaning

Similarly with the other dataframe, let's:
*   Lowercase everything
*   Remove numbers and punctuation
*   Remove all words except nouns
*   Perform lemmatization (more computationally expensive, but preserves real words)
*   Remove random one letter words

```python
thousand_books['description_cleaned'] = thousand_books['description'].apply(lambda x: x.lower() if isinstance(x, str) else x)
thousand_books
```

```python
!pip install nltk
import nltk

nltk.download('punkt')

from nltk import word_tokenize, pos_tag
nltk.download('averaged_perceptron_tagger')
```

```python
def filter_nouns(text):
    words = word_tokenize(text)  # Tokenize the text into words
    tagged_words = pos_tag(words)  # Perform part-of-speech tagging

    # Keep only nouns (NN, NNS, NNP, NNPS)
    filtered_words = [word for word, tag in tagged_words if tag.startswith('NN')]

    filtered_text = ' '.join(filtered_words)  # Join the filtered words back into a string
    return filtered_text
```

```python
# get rid of all words that aren't nouns

%%time
thousand_books['description_cleaned'] = thousand_books['description_cleaned'].apply(lambda x: filter_nouns(x) if isinstance(x, str) else x)
thousand_books
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
%%time
thousand_books['description_cleaned'] = thousand_books['description_cleaned'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]) if isinstance(x, str) else x)
thousand_books
```

```python
import re
thousand_books['description_cleaned'] = thousand_books['description_cleaned'].apply(lambda x: re.sub(r'[^\w\s]', ' ', x) if isinstance(x, str) else x)
thousand_books
```

```python
def remove_one_letter(text):
  words = text.split()
  filtered_words = [word for word in words if len(word)>1]
  return ' '.join(filtered_words)

thousand_books['description_cleaned'] = thousand_books['description_cleaned'].apply(remove_one_letter)
```
